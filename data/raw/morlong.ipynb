## รายละเอียดการทดลอง

การทดลองนี้เป็นการเก็บข้อมูลจาก Sensor ของหุ่นยนต์ RoboMaster EP 
โดยทำการบันทึกข้อมูลจาก Sensor ต่าง ๆ ได้แก่

- Attitude Sensor สำหรับวัดค่ามุม Orientation ของหุ่นยนต์
- ESC Sensor สำหรับตรวจสอบค่ามอเตอร์และความเร็วล้อ
- IMU Sensor สำหรับวัดความเร่งและการหมุนของตัวหุ่นยนต์
- Position Sensor สำหรับติดตามตำแหน่งของหุ่นยนต์
- Chassis Status สำหรับตรวจสอบสถานะของฐานล้อ
- Gyroscope Sensor สำหรับวัดความเร็วเชิงมุม

จากนั้นนำข้อมูลที่บันทึกไว้ในไฟล์ CSV มาทำการ Plot กราฟ
เพื่อวิเคราะห์พฤติกรรมการเคลื่อนที่ของหุ่นยนต์

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
files = {
    "attitude_log":"attitude_log.csv",
    "esc_log":"esc_log.csv",
    "imu_log":"imu_log.csv",
    "position_log":"position_log.csv",
    "chassis_status_log":"chassis_status_log.csv",
    "gyro_log":"gyro_log.csv"
}


data = {}

for name,file in files.items():
    try:
        data[name] = pd.read_csv(file)
        print(name,"โหลดสำเร็จ")
        display(data[name].head())

    except:
        print(name,"ไม่พบไฟล์")
```

## Plot attitude_log

```python
df = data["attitude_log"]

plt.figure(figsize=(12,5))

for col in df.columns:
    if col.lower() != "time":
        plt.plot(df[col],label=col)

plt.title("attitude_log Sensor")
plt.xlabel("Sample")
plt.ylabel("Value")

plt.legend()
plt.grid()

plt.show()
```

## Plot esc_log

```python
df = data["esc_log"]

plt.figure(figsize=(12,5))

for col in df.columns:
    if col.lower() != "time":
        plt.plot(df[col],label=col)

plt.title("esc_log Sensor")

plt.xlabel("Sample")
plt.ylabel("Value")

plt.legend()
plt.grid()

plt.show()
```
