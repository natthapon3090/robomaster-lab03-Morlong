```python
import pandas as pd
import matplotlib.pyplot as plt

# --- ส่วนที่ 1: ดึงไฟล์ CSV ของเซนเซอร์ IMU มาพลอตกราฟแยกแผง (Lab 1) ---
df_imu = pd.read_csv('../data/log_imu.csv')

# ปรับค่าแกน X ลำดับเวลาให้เริ่มต้นคำนวณที่วินาทีที่ 0
time_steps = (df_imu['timestamp'] - df_imu['timestamp'].iloc[0]) / 1000.0

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), sharex=True)

# กราฟความเร่งแกน X และ Y
ax1.plot(time_steps, df_imu['acc_x'], 'r-', label='Acceleration X')
ax1.plot(time_steps, df_imu['acc_y'], 'g-', label='Acceleration Y')
ax1.set_ylabel('ความเร่ง (m/s²)')
ax1.grid(True)
ax1.legend()

# กราฟความเร็วเชิงมุมแกน Z (รอบการเลี้ยว/ความนิ่งขณะสไลด์)
ax2.plot(time_steps, df_imu['gyro_z'], 'b-', label='Gyro Z')
ax2.set_ylabel('ความเร็วเชิงมุม แกน Z (deg/s)')
ax2.set_xlabel('เวลาหน่วย (วินาที)')
ax2.grid(True)
ax2.legend()

plt.suptitle("กราฟแสดงผลการล็อกข้อมูลเซนเซอร์ IMU แยกเป็นแผงย่อย")
plt.show()

# --- ส่วนที่ 2: ดึงไฟล์คำนวณวัดค่าความหน่วงการรับส่งข้อมูล (Lab 2) ---
df_lat = pd.read_csv('../data/log_latency.csv')
avg_latency = df_lat['latency_ms'].mean()
print(f"ค่าความหน่วงเฉลี่ยของการรับส่งข้อมูล (Average Latency): {avg_latency:.2f} มิลลิวินาที (ms)")
```
